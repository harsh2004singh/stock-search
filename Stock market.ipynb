```python
from bsedata.bse import BSE

def get_real_time_stock_price(stock_code):
    bse = BSE(update_codes=True)  # Initialize BSE object with update_codes = True for real-time data
    stock_data = bse.getQuote(stock_code)
    if stock_data:
        return stock_data['currentValue']
    else:
        return None

def main():
    stock_code = '500209'  # Example: TCS stock code
    real_time_price = get_real_time_stock_price(stock_code)
    if real_time_price:
        print(f"Real-time price for stock with code {stock_code}: {real_time_price}")
    else:
        print(f"Unable to fetch real-time price for stock with code {stock_code}")

if __name__ == "__main__":
    main()
```
